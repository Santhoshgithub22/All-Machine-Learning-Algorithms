```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
X1, y1 = make_classification(n_samples=1000, n_features=2, n_redundant = 0,
                    n_informative=2, n_clusters_per_class=1,
                    class_sep=2.0, random_state=42)
```

```python
y1
```

```python
plt.scatter(X1[:,0], X1[:,1], c=y1)
plt.show()
```

#### Now lets see about one more dataset

```python
from sklearn.datasets import make_moons

X3, y3 = make_moons(n_samples=1000, noise=0.3, random_state=42)
```

```python
X3
```

```python
y3
```

```python
plt.scatter(X3[:,0], X3[:,1], c=y3)
plt.show() # This is not linearly separable, kernal should be use here
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Model Building

from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred) # I think overfitting so check train accuracy too
```

```python
clf.score(X_train, y_train) # Its near to each other, so my model is good
```

# Now doing model with other data

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

```python
# Model Building

from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train3, y_train3)
```

```python
clf.score(X_train3, y_train3)
```

```python
y_pred3 = clf.predict(X_test3)
```

```python
accuracy_score(y_test3, y_pred3)
```
